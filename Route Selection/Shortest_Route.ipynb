# ***Importing Libraries***



---



```python
import folium
import osmnx as ox
import networkx as nx
from geopy.geocoders import Nominatim
```

# ***Initializng***

---



```python
locator = Nominatim(user_agent = "myapp")
```

```python
ox.config(log_console=True, use_cache=True)
```

# ***Specifying Starting Location***


---



```python
start_location = "Brisbane City Botanic Gardens"
```

# ***Specifying Destination Location***


---

```python
end_location = "Riverside Sunday Market"
```

# ***Attaining Co-ordinates***



---



```python
start_latlng = locator.geocode(start_location).point
```

```python
start_latlng
```

```python
end_latlng = locator.geocode(end_location).point
```

```python
end_latlng
```

# ***Location where to find route***



---



```python
place = 'Brisbane, Queensland, Australia'
```

# ***Mode of Travel***



---



```python
# find shortest route based on the mode of travel
mode = 'drive'       # 'drive', 'bike', 'walk'
```

# ***Selecting Optimizer***


---



```python
optimizer = 'time' 
```

# ***Finding the shortest route***



---



```python
# create graph from OSM within the boundaries of some geocodable place(s)
graph = ox.graph_from_place(place, network_type = mode)
```

```python
graph
```

```python
# find the nearest node to the start location
orig_node = ox.get_nearest_node(graph, start_latlng)
```

```python
orig_node
```

```python
# find the nearest node to the end location
dest_node = ox.get_nearest_node(graph, end_latlng)
```

```python
dest_node 
```

```python
#  find the shortest path using dijkstra
shortest_route = nx.shortest_path(graph, orig_node, dest_node, weight=optimizer)
```

```python
shortest_route
```

# ***Plotting Routes***



---



```python
shortest_route_map = ox.plot_route_folium(graph, shortest_route)
shortest_route_map
```

```python
shortest_route_map = ox.plot_route_folium(graph, shortest_route, tiles='openstreetmap')
shortest_route_map
```

```python
folium.TileLayer('openstreetmap').add_to(shortest_route_map)
folium.TileLayer('Stamen Terrain').add_to(shortest_route_map)
folium.TileLayer('Stamen Toner').add_to(shortest_route_map)
folium.TileLayer('Stamen Water Color').add_to(shortest_route_map)
folium.TileLayer('cartodbpositron').add_to(shortest_route_map)
folium.TileLayer('cartodbdark_matter').add_to(shortest_route_map)

folium.LayerControl().add_to(shortest_route_map)

shortest_route_map
```
